```python
from ft_linear_regression import FtLinearRegression as LR
import numpy as np
import matplotlib.pyplot as plt
```

```python
lg = LR(learning_rate=0.05, n_iterations=1000)

print("Mostrar os valores iniciais de theta0 e theta1")
print(f"theta0: {lg.theta0}, theta1: {lg.theta1}")
print(f"Previsão com tudo zerado: {lg.predict(200)}")
print(f"Variaveis da classe: {lg.X}, {lg.y}, {lg.X_mean}, {lg.y_mean}")
```

```python
print("Ler os dados do arquivo CSV")
data = lg.read_csv("data.csv")
print(f"Dados lidos: {data}\n{len(data)} linhas")
```

```python
lg.fit_intern_data()
print("theta0 e theta1 após o treinamento")
print(f"theta0: {lg.theta0}, theta1: {lg.theta1}")
print(f"Previsão com os dados treinados: {lg.predict(200)}")
print(f"Variaveis da classe: X mean {lg.X_mean:.2f}, X std {lg.X_std:.2f}, y mean {lg.y_mean:.2f}, y std {lg.y_std:.2f}")
```

```python
x = [row[0] for row in data]
predictions = lg.predict(x)
r2 = lg.r2_score([row[1] for row in data], predictions)

print("Previsões e R² score com dados ja treinados")
print(f"Variaveis da classe:")
print(f"X mean: {lg.X_mean:.2f}, X std: {lg.X_std:.2f}, y mean: {lg.y_mean:.2f}, y std: {lg.y_std:.2f}")
print(f"theta0: {lg.theta0}, theta1: {lg.theta1}")
print(f"R² score do modelo: {r2}")

plt.scatter(x, [row[1] for row in data], color='blue', label='Dados reais')
plt.plot(x, predictions, color='red', label='Previsão')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
x = [row[0] for row in data]
# Gerar novos dados de x fora do intervalo original para testar previsões em dados desconhecidos
# mas com o mesmo treinamento anterior
x = np.linspace(min(x) * 0.8, max(x) * 1.2, len(x))

predictions = lg.predict(x)
r2 = lg.r2_score([row[1] for row in data], predictions)
print("Previsões e R² score com dados desconhecidos")
print(f"Variaveis da classe:")
print(f"X mean: {lg.X_mean:.2f}, X std: {lg.X_std:.2f}, y mean: {lg.y_mean:.2f}, y std: {lg.y_std:.2f}")
print(f"theta0: {lg.theta0}, theta1: {lg.theta1}")
print(f"R² score do modelo: {r2}")

plt.scatter(x, [row[1] for row in data], color='blue', label='Dados reais')
plt.plot(x, predictions, color='red', label='Previsão')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# criar novos dados, treinar com fit e prever novamente
new_data = [[i, 2 * i + 3 + np.random.normal(0, 1)] for i in range(25)]
middle_index = len(new_data) // 2
X_fit = [row[0] for row in new_data[:middle_index]]
y_fit = [row[1] for row in new_data[:middle_index]]
lg.fit(X_fit, y_fit)

X_predict = [row[0] for row in new_data[middle_index:]]
y_predict = [row[1] for row in new_data[middle_index:]]
print("theta0 e theta1 após o novo treinamento")
print(f"theta0: {lg.theta0}, theta1: {lg.theta1}")
predictions = lg.predict(X_predict)

r2 = lg.r2_score(y_predict, predictions)
print("Previsões e R² score com dados desconhecidos")
print(f"Variaveis da classe:")
print(f"X mean: {lg.X_mean:.2f}, X std: {lg.X_std:.2f}, y mean: {lg.y_mean:.2f}, y std: {lg.y_std:.2f}")
print(f"theta0: {lg.theta0}, theta1: {lg.theta1}")
print(f"R² score do modelo: {r2}")

plt.scatter(X_fit + X_predict, y_fit + y_predict, color='blue', label='Dados de treinamento')
plt.plot(X_predict, predictions, color='red', label='Previsão')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
